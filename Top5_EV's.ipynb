```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('cleaned_ev_scooter_sales.csv')

# Display first 5 rows
print("Data Preview:")
display(df.head())


# %% [markdown]
# ## 4. Basic Analysis
# ### Summary Statistics
print("\nSummary Statistics:")
display(df.describe())

# ### Top 5 Models by Sales (Apr-25)
top_models = df.sort_values('Apr-25', ascending=False).head(5)
print("\nTop 5 Models by Sales (Apr-25):")
display(top_models[['OEM', 'Model', 'Apr-25']])

# ### Worst 5 Models by YoY Growth
worst_models = df.sort_values('YoY Growth (%)').head(5)
print("\nWorst 5 Models by YoY Growth:")
display(worst_models[['OEM', 'Model', 'YoY Growth (%)']])

# %% [markdown]
# ## 5. Visualizations
# ### Bar Plot: Top 5 Models by Sales
plt.figure(figsize=(10, 5))
sns.barplot(data=top_models, x='Model', y='Apr-25', hue='OEM')
plt.title('Top 5 EV Scooters by Sales (Apr-2025)')
plt.xticks(rotation=45)
plt.show()


```
